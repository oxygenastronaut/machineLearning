# Learning Pandas with Pokemon Stats

### taken from https://www.kaggle.com/ash316/learn-pandas-with-pokemons

```python
# importing all packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the data
df = pd.read_csv('Pokemon.csv')
df.head(10)
```

```python
# changing the columns to upper case
df.columns = df.columns.str.upper().str.replace('_', '')
df.head(10)
```

### To show the legendary pokemon we can do this

```python
df[df['LEGENDARY'] == True].head(10)
```

```python
df = df.set_index('NAME') #change and set the index to the
# name attribute
```

# Cleaning the Dataframe

```python
## The index of Mega Pokemon contain extra text. 
df.index = df.index.str.replace(".*(?=Mega)", '')
df.head(10)
```

```python
df = df.drop(['#'], axis = 1) # drop the columns with axis = 1; axis = 0 is for rows
```

```python
print('The columns of the dataset are: ', df.columns) 
print('The shape of the dataframe is: ', df.shape)
```

```python
# some values in Type 2 are empty so to fill them
df['TYPE 2'].fillna(df['TYPE 1'], inplace= True) # fill NaN into Type2 with corresponding values of Type1
df.head(10)
```

```python
print(df.loc['Bulbasaur']) # retrieves complete row data from index with value Bulbasaur
print(df.iloc[0]) # retrieves complete row data from index 0; integer version of loc
print(df.ix[0]) # similar to iloc
print(df.ix['Kakuna']) # similar to loc
# loc works on labels in the index
# iloc works on positions in the index so it only takes integers
# ix usually tries to behave like loc but falls back to behaving like iloc if the labels is not in the index
# inoreder to find details about any pokemon, just specify its name
```

```python
# filtering pokemon using logical operators
df[((df['TYPE 1'] == 'Fire') | (df['TYPE 1'] == 'Dragon')) & 
   ((df['TYPE 2'] == 'Fire') | (df['TYPE 2'] == 'Dragon'))].head(3)
```

```python
print('Max HP: ', df['HP'].argmax()) # returns the pokemon with the highest HP stat
print('Max Denfense: ', df['DEFENSE'].idxmax()) # similar to argmax
```

```python
df.sort_values('TOTAL', ascending = False).head(5) # this arranges in descending order
# ascending = True would make it go in reverse order to False
```

```python
print(df['TYPE 1'].value_counts(), '\n', df['TYPE 2']. value_counts()) # counts different types
df.groupby(['TYPE 1']).size() # same as above
(df['TYPE 1']== 'Bug').sum() # counts for a single value
```

```python
df[(df['TYPE 1']=='Flying')].head() # there are actually 4 primary flying types. cool
```

```python
df_summary = df.describe() # summary of the dataframe
df_summary
```

```python
fire = df[(df['TYPE 1']=='Fire') | (df['TYPE 2'] == 'Fire')] # contains all fire pokemon
water = df[(df['TYPE 1']=='Water') | (df['TYPE 2'] == 'Water')] # all water 
plt.scatter(fire.ATTACK.head(50), fire.DEFENSE.head(50), color = 'R', label = 'Fire', marker = '*', s = 50)
plt.scatter(water.ATTACK.head(50), water.DEFENSE.head(50), color = 'B', label = 'Water', s = 25)
plt.xlabel('Attack')
plt.ylabel('Defense')
```
